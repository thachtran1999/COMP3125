```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df_1 = pd.read_csv("C:/Users/thach/Downloads/indiv_proj/sp500_stocks.csv")
```

```python
df_2 = pd.read_csv("C:/Users/thach/Downloads/indiv_proj/sp500_companies.csv")[['Symbol', 'Sector','Longname']]
```

```python
df = pd.merge(df_1, df_2, on='Symbol', how='left')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
# Question 1:
sector_avg = df.groupby(['Sector', 'Date'])['Close'].mean()

pct_change = ((sector_avg.groupby(level=0).last() / sector_avg.groupby(level=0).first() - 1) * 100)

all_sectors = pct_change.reset_index().rename(columns={'Close':'Percentage Change'})
all_sectors = all_sectors.sort_values('Percentage Change', ascending=False)
all_sectors.index = range(1, len(all_sectors)+1)

print("Performance of all sectors:")
print(all_sectors)
```

```python
# Question 2:
top_sector = 'Communication Services'
df_top = df[df['Sector'] == top_sector]

# Compute average close by date
sector_trend = df_top.groupby('Date')['Close'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Close', data=sector_trend, linewidth=2.5, marker=None)
plt.title(f'{top_sector} Sector - Average Closing Price Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Close Price', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Question 3:
stability = df_top.groupby('Symbol')['Close'].std().reset_index()
stability = stability.rename(columns={'Close':'Standard Deviation'})

symbol_to_name = df_top[['Symbol', 'Longname']].drop_duplicates('Symbol').set_index('Symbol')['Longname']

stability['Longname'] = stability['Symbol'].map(symbol_to_name)

stability = stability[['Symbol', 'Longname', 'Standard Deviation']]

stability_sorted = stability.sort_values('Standard Deviation').reset_index(drop=True)
stability_sorted.index = stability_sorted.index + 1

stability_sorted['Standard Deviation'] = stability_sorted['Standard Deviation'].round(2)

print("Top 10 Most Stable Companies in", top_sector)
print(stability_sorted.head(10))
```

```python
# Question 4:
company = 'NVDA'
df_company = df[df['Symbol'] == company].sort_values('Date')
df_company = df_company[df_company['Close'].notna()]

df_recent = df_company.tail(30).copy()

df_recent['Close_lag1'] = df_recent['Close'].shift(1)
df_recent['Close_lag2'] = df_recent['Close'].shift(2)
df_recent['MA_5'] = df_recent['Close'].rolling(5).mean()
df_recent['MA_10'] = df_recent['Close'].rolling(10).mean()
df_recent['MA_20'] = df_recent['Close'].rolling(20).mean()
df_recent = df_recent.dropna()

features = ['Open', 'High', 'Low', 'Volume', 'Close_lag1', 'Close_lag2', 'MA_5', 'MA_10', 'MA_20']
X = df_recent[features]
y = df_recent['Close']

reg_model = LinearRegression()
reg_model.fit(X, y)

forecast_horizon = 5
forecast_reg = []
last_features = X.iloc[[-1]].copy()
recent_closes = list(df_recent['Close'].values)

for i in range(forecast_horizon):
    pred = reg_model.predict(last_features)[0]
    forecast_reg.append(pred)
    
    last_features['Close_lag2'] = last_features['Close_lag1']
    last_features['Close_lag1'] = pred
    
    recent_closes.append(pred)
    last_features['MA_5'] = np.mean(recent_closes[-5:])
    last_features['MA_10'] = np.mean(recent_closes[-10:])
    last_features['MA_20'] = np.mean(recent_closes[-20:])

ets_model = ExponentialSmoothing(df_recent['Close'], trend='add', seasonal=None)
fit = ets_model.fit()
forecast_ets = fit.forecast(forecast_horizon)

combined_forecast = 0.5 * np.array(forecast_reg) + 0.5 * np.array(forecast_ets)
forecast_dates = pd.date_range(df_recent['Date'].iloc[-1] + pd.Timedelta(days=1), periods=forecast_horizon)

predicted_next = combined_forecast[0]
print(f"Predicted closing prices for {company} for the next {forecast_horizon} trading days:")
for date, price in zip(forecast_dates, combined_forecast):
    print(f"{date.date()}: ${price:.2f}")
    
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Close', data=df_recent, marker='o', label='Actual')
sns.lineplot(x=forecast_dates, y=combined_forecast, color='red', linestyle='--', marker='o', label='Forecast')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.title(f"{company} Actual & Forecasted Closing Prices (Combined Regression + ETS)")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.tight_layout()
plt.show()
```
